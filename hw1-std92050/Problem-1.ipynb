# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def traj(step,dx):
    x=np.array([])
    for j in range(step):
        x=np.append(x,np.sum(dx[0:j]))
    plt.plot([i for i in range(step)],x)

    
def disc(x):
    if x>=(1/2):
        return 1
    else:
        return -1

def hist_uni(walker,step):
    S=np.array([])
    for i in range(walker):
        dx=np.array([])
        a=np.random.random((step,))
        for b in a:
            dx=np.append(dx,b-(1/2))
        S=np.append(S,np.sum(dx))
    plt.hist(S,bins=50)

def hist_disc(walker,step):
    S=np.array([])
    for i in range(walker):
        dx=np.array([])
        a=np.random.random((step,))
        for b in a:
            dx=np.append(dx,disc(b))
        S=np.append(S,np.sum(dx))
    plt.hist(S,bins=50)
    




step=1000
walker=1000
```

```python
#uniform distribution on an interval
#trajecctory
dx=np.random.random((step,))-(1/2)
traj(step,dx)
```

```python
#uniform distribution on an interval
#histgram
hist_uni(walker,step)
```

```python
#discrete traj
dx=np.array([])
a=np.random.random((step,))
for b in a:
    dx=np.append(dx,disc(b))
traj(step,dx)
```

```python
#disc hist
hist_disc(walker,step)
```
