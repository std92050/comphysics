```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.random((1000,))
```

```python
fix,ax= plt.subplots()
_=ax.hist(x,bins=50)
```

```python
plt.scatter([i for i in range(1000)],x)
```

```python
#1D walk
step=100000
walker=100
S=np.array([])
for i in range(walker-1):
    dx=2*np.random.random((step,))-1
    
    S=np.append(S,np.sum(dx))
plt.scatter([i for i in range(99)],S)
```

```python
plt.hist(S,bins=50)
```

```python
#2d walk
step=1000
walker=100
Sx=np.array([])
Sy=np.array([])
for i in range(walker-1):
    dx=2*np.random.random((step,))-1
    dy=np.random.random((step,))-(1/2)
    Sx=np.append(Sx,np.sum(dx))
    Sy=np.append(Sy,np.sum(dy))

plt.scatter(Sx,Sy)
```
